```python
import requests
import json
import numpy as np

def check_etf_availability(etf_name, investengine_url="https://investengine.com/api/v0.29/public/securities/"):
    """
    Checks if an ETF is available on InvestEngine by searching their API.

    Args:
        etf_name (str): The name or ticker of the ETF to search for (e.g., "Amundi MSCI").
        investengine_url (str, optional): The URL of the InvestEngine API endpoint.
            Defaults to the InvestEngine securities API.

    Returns:
        str: A message indicating whether the ETF is available or not, or if an error occurred.
    """
    try:
        # 1. Construct the InvestEngine API search URL.
        search_url = f"{investengine_url}?search={etf_name.replace(' ', '%20')}"  # Spaces to %20 for URLs
        print(f"Searching URL: {search_url}")

        # 2. Send the GET request to the InvestEngine API.
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
        response = requests.get(search_url, headers=headers)
        response.raise_for_status()  # Raise an exception for bad status codes

        # 3. Parse the JSON response.
        if response.status_code == 304:
            return f"ETF '{etf_name}' was not modified since last check." #Handles 304
        json_response = response.json()

        # 4. Check for results in the response.
        if isinstance(json_response, list) and len(json_response) > 0: #checks the type
            first_hit = json_response[0]
            #Adapt the keys.  Uses the keys from the provided JSON.
            etf_title = first_hit.get('title', 'N/A')
            etf_ticker = first_hit.get('ticker', 'N/A')
            return True#f"ETF '{etf_name}' is available on InvestEngine. Details: Title: {etf_title}, Ticker: {etf_ticker}"
        else:
            return False#f"ETF '{etf_name}' does not appear to be available on InvestEngine."

    except requests.exceptions.RequestException as e:
        return f"Error: {e}"
    except json.JSONDecodeError as e:
        return f"Error decoding JSON response: {e}. Response text: {response.text}"
    except Exception as e:
        return f"An unexpected error occurred: {e}"

#for each ticker in main_df, check if the ETF is available on InvestEngine and add the result to a new column
# Create a list to store the results
investor_engine_results = []

# Iterate over each row in the DataFrame
for index, row in main_df.iterrows():
    etf_name = row['ticker']
    result = check_etf_availability(etf_name)
    investor_engine_results.append(result)

#remove investengine from the main_df
main_df = main_df.drop(columns=['investengine'], errors='ignore')
# Add the new column at the beginning of the DataFrame
main_df.insert(0, 'investengine', investor_engine_results)  # Insert at position 0 (beginning)

# Save the final DataFrame to a CSV file
output_file = "final.csv"
main_df.to_csv(output_file, index=False)

# Select the specified columns from main_df
selected_columns = [
    'Country', '2023 GDP', 'MSCI', 'Region', 'Short_name', 'ticker', 'name', 'investengine',
    'age_in_years', 'domicile_country', 'currency', 'dividends', 'ter', 
    'replication', 'size', 'yesterday', '2024', '2023', '2022', '2021', 
    'last_year_dividends', 'last_year_volatility', 'last_year_max_drawdown', 
    'last_year_return_per_risk'
]

# Create a new DataFrame with the selected columns, filter ones with investengine == True, 
# and if there are 2 funds in the same country show the one with lower ter, higher dividend last year and higher risk return for last year
new_df = main_df[(main_df['investengine'] == True and main_df["Asset_Class"]== "Equities") or main_df["Asset_Class"]== "Bonds"]
#new_df = new_df[selected_columns]
new_df = new_df.sort_values(by=['Country', 'ter', 'last_year_dividends', 'last_year_return_per_risk'], 
                              ascending=[True, True, False, False])
# Drop duplicates based on 'Country' and keep the first occurrence
new_df = new_df.drop_duplicates(subset=['Country'], keep='first')

##Allocate starting risk weights - Equities 90% and Bonds 10%
trailing_sr_adjustment_factors = {
    -1: 0.6,
    -0.8: 0.66,
    -0.6: 0.77,
    -0.4: 0.85,
    -0.2: 0.94,
    0: 1,
    0.2: 1.11,
    0.4: 1.19,
    0.6: 1.3,
    0.8: 1.37,
    1: 1.48
}
eq_risk_weight = 0.9
bnd_risk_weight = 0.1
eq_benchmark_12mosreturn = 0.12 # TODO: get this from the data
bnd_benchmark_12mosreturn = -0.04 # TODO: get this from the data
eq_benchmark_volatility = 0.2 # TODO: get this from the data
bnd_benchmark_volatility = 0.05 # TODO: get this from the data
eq_sharp_ratio = eq_benchmark_12mosreturn / eq_benchmark_volatility
bond_sharp_ratio = bnd_benchmark_12mosreturn / bnd_benchmark_volatility
#calculate the trailing sharp ratio for eq and bond. 
# If sharp ratio is less than -1, use 0.6, if sharp ration is greater than 1, use 1.48, 
# else use the value from the dictionary
# Extract keys and values from the dictionary
keys = np.array(list(trailing_sr_adjustment_factors.keys()))  # J5:J15
values = np.array(list(trailing_sr_adjustment_factors.values()))  # I5:I15


# Implement the logic
if eq_sharp_ratio <= values[0]:
    eq_interpolated_adj_factor = keys[0]
elif eq_sharp_ratio >= values[-1]:
    eq_interpolated_adj_factor = keys[-1]
else:
    # Perform linear interpolation
    eq_interpolated_adj_factor = np.interp(eq_sharp_ratio, values, keys)

#calculate bond_interpolated_adj_factor
if bond_sharp_ratio <= values[0]:
    bond_interpolated_adj_factor = keys[0]
elif bond_sharp_ratio >= values[-1]:
    bond_interpolated_adj_factor = keys[-1]
else:
    # Perform linear interpolation
    bond_interpolated_adj_factor = np.interp(bond_sharp_ratio, values, keys)

eq_adj_risk_weights = eq_risk_weight * eq_interpolated_adj_factor
bond_adj_risk_weights = bnd_risk_weight * bond_interpolated_adj_factor
normalized_eq_adj_risk_weights = eq_adj_risk_weights / (eq_adj_risk_weights + bond_adj_risk_weights)
normalized_bond_adj_risk_weights = bond_adj_risk_weights / (eq_adj_risk_weights + bond_adj_risk_weights)
eq_cash_weights = eq_adj_risk_weights/eq_benchmark_volatility # inverse volatility weighting
bond_cash_weights = bond_adj_risk_weights/bnd_benchmark_volatility
final_eq_cash_weights = normalized_eq_adj_risk_weights * (eq_cash_weights / (eq_cash_weights + bond_cash_weights))
final_bond_cash_weights = normalized_bond_adj_risk_weights * (bond_cash_weights / (eq_cash_weights + bond_cash_weights))
print(f"Final Equities Cash Weights: {final_eq_cash_weights}")
print(f"Final Bonds Cash Weights: {final_bond_cash_weights}")


print(f"Interpolated/Extrapolated Result: {result}")



# allocate a weighted score to each fund 
# 0.2*last_5_years_return_per_risk + 0.3*last_3_years_return_per_risk + 0.5*last_year_return_per_risk
new_df['weighted_score'] = (0.2 * new_df['last_five_years_return_per_risk'] +
                            0.3 * new_df['last_three_years_return_per_risk'] +
                            0.5 * new_df['last_year_return_per_risk'])
#normalize the score so that the total score is 1
new_df['weighted_score'] = new_df['weighted_score'] / new_df['weighted_score'].sum()
#invest 10000 across all the funds
new_df['investment'] = round(abs(new_df['weighted_score'] * 10000),2)

output_mini_file = "final_mini.csv"
new_df.to_csv(output_mini_file, index=False)
```

```python
import requests
import pandas as pd

def calculate_etf_metrics(symbol, custom_start_end=None):
    """
    Calculate annual and monthly metrics for a given ETF symbol, with optional custom start and end prices.

    Args:
        symbol (str): The ticker symbol of the ETF (e.g., 'CACX.LON').
        custom_start_end (dict, optional): A dictionary specifying custom start and end prices for specific years or months.
            Example: {"2024": {"start_price": 100, "end_price": 110}}

    Returns:
        pd.DataFrame: A DataFrame containing the annual metrics.
    """
    # Replace the "demo" API key below with your own key from https://www.alphavantage.co/support/#api-key
    url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol={symbol}&outputsize=full&apikey=1JYCUCSEKO7IYQKU'

    r = requests.get(url)
    data = r.json()
    df = pd.DataFrame(data['Time Series (Daily)']).T

    # Convert index to datetime
    df.index = pd.to_datetime(df.index)

    # Convert relevant columns to numeric and convert prices to pounds
    df['5. adjusted close'] = pd.to_numeric(df['5. adjusted close'], errors='coerce') / 100  # Convert to pounds

    # Round the raw data to 2 decimals
    df['5. adjusted close'] = df['5. adjusted close'].round(2)

    # Define the years to calculate returns for
    years = [2021, 2022, 2023, 2024, 2025]

    # Create a list to store annual results
    annual_results = []

    # Create a list to store monthly results
    monthly_results = []

    # Loop through each year and calculate metrics
    for year in years:
        # Filter the data for the current year
        df_year = df[df.index.year == year]
        
        if df_year.empty:
            print(f"No data available for {year}.")
            continue

        # Check if custom start and end prices are provided for the year
        if custom_start_end and str(year) in custom_start_end:
            start_price = custom_start_end[str(year)].get("start_price", df_year['5. adjusted close'].iloc[-1])
            end_price = custom_start_end[str(year)].get("end_price", df_year['5. adjusted close'].iloc[0])
        else:
            # Default to calculating start and end prices from the data
            start_price = df_year['5. adjusted close'].iloc[-1]  # Last row of the year (start of the year)
            end_price = df_year['5. adjusted close'].iloc[0]    # First row of the year (end of the year)

        # Calculate the total return for the year
        total_return = ((end_price) / start_price) - 1

        # Since we are calculating for a single year, the annualized return is the same as the total return
        annualized_return = total_return

        # Calculate volatility (standard deviation of daily returns)
        df_year['daily_return'] = df_year['5. adjusted close'].pct_change()
        volatility = df_year['daily_return'].std() * (252 ** 0.5)  # Annualized volatility

        # Append the annual results to the list (rounded to 2 decimal places)
        annual_results.append({
            "Year": year,
            "Start Price": round(start_price, 2),
            "End Price": round(end_price, 2),
            "Annualized Return (%)": round(annualized_return * 100, 2),
            "Volatility (%)": round(volatility * 100, 2)
        })

        # Group the data by month and calculate monthly metrics
        df_year['month'] = df_year.index.to_period('M')
        for month, df_month in df_year.groupby('month'):
            # Check if custom start and end prices are provided for the month
            month_str = month.strftime('%Y-%m')
            if custom_start_end and month_str in custom_start_end:
                start_price_month = custom_start_end[month_str].get("start_price", df_month['5. adjusted close'].iloc[-1])
                end_price_month = custom_start_end[month_str].get("end_price", df_month['5. adjusted close'].iloc[0])
            else:
                # Default to calculating start and end prices from the data
                start_price_month = df_month['5. adjusted close'].iloc[-1]  # Last row of the month (start of the month)
                end_price_month = df_month['5. adjusted close'].iloc[0]    # First row of the month (end of the month)

            # Calculate the monthly return
            monthly_return = ((end_price_month) / start_price_month) - 1

            # Calculate monthly volatility (standard deviation of daily returns)
            monthly_volatility = df_month['daily_return'].std() * (252 ** 0.5)  # Annualized volatility

            # Append the monthly results to the list (rounded to 2 decimal places)
            monthly_results.append({
                "Year": year,
                "Month": month_str,
                "Start Price": round(start_price_month, 2),
                "End Price": round(end_price_month, 2),
                "Monthly Return (%)": round(monthly_return * 100, 2),
                "Volatility (%)": round(monthly_volatility * 100, 2)
            })

    # Create DataFrames from the results
    annual_results_df = pd.DataFrame(annual_results)
    monthly_results_df = pd.DataFrame(monthly_results)

    # Print the annual results as a table
    print("\nAnnual Metrics:")
    print(annual_results_df.to_string(index=False))

    # Print the monthly results as a table
    print("\nMonthly Metrics:")
    print(monthly_results_df.to_string(index=False))

    # Return the annual metrics DataFrame
    return annual_results_df

# Example usage
symbol = "CACX.LON"
custom_prices = {}
"""    "2021": {"start_price": 87.81, "end_price": 107.99},  # Custom prices in pounds
    "2022": {"start_price": 107.99, "end_price": 106.12},
    "2023": {"start_price": 106.12, "end_price": 124.66},
    "2024": {"start_price": 125.29, "end_price": 119.75}
}"""
annual_metrics = calculate_etf_metrics(symbol, custom_start_end=custom_prices)
```

```python
import pandas as pd
import requests

# Assuming calculate_etf_metrics is already defined in your script
# Function to calculate the 2024 metrics for a given ticker
def get_2024_metrics(ticker, ter):
    """
    Calculate the 2024 metrics (returns and Sharpe ratio) for a given ETF ticker.

    Args:
        ticker (str): The ticker symbol of the ETF (e.g., 'CACX.LON').
        ter (float): The Total Expense Ratio (TER) for the ETF.

    Returns:
        dict: A dictionary containing the 2024 returns and Sharpe ratio.
    """
    try:
        # Call the calculate_etf_metrics function
        annual_metrics = calculate_etf_metrics(ticker)
        
        # Filter for the 2024 row
        metrics_2024 = annual_metrics.loc[annual_metrics['Year'] == 2024]
        
        if not metrics_2024.empty:
            # Extract relevant metrics
            returns_2024 = metrics_2024['Annualized Return (%)'].values[0]
            yield_2024 = metrics_2024['Yield (%)'].values[0]
            volatility_2024 = metrics_2024['Volatility (%)'].values[0]
            
            # Calculate net yield and Sharpe ratio
            net_yield = yield_2024 - ter
            sharpe_ratio = net_yield / volatility_2024 if volatility_2024 != 0 else None
            
            return {
                "2024_returns": round(returns_2024, 2),
                "2024_sharpe_ratio": round(sharpe_ratio, 2) if sharpe_ratio is not None else None
            }
        else:
            return {
                "2024_returns": None,
                "2024_sharpe_ratio": None
            }
    except Exception as e:
        print(f"Error calculating metrics for {ticker}: {e}")
        return {
            "2024_returns": None,
            "2024_sharpe_ratio": None
        }

# Load the final_mini.csv file
final_mini_file = "final_mini.csv"
final_mini_df = pd.read_csv(final_mini_file)

# Add the .LON suffix to the ticker column
final_mini_df['ticker_with_suffix'] = final_mini_df['ticker'] + '.LON'

# Calculate the 2024 metrics for each ETF
metrics_results = []
for index, row in final_mini_df.iterrows():
    ticker = row['ticker_with_suffix']
    ter = row['ter']  # Use the TER value from the DataFrame
    metrics = get_2024_metrics(ticker, ter)
    metrics_results.append(metrics)

# Add the new metrics to the DataFrame
metrics_df = pd.DataFrame(metrics_results)
final_mini_df = pd.concat([final_mini_df, metrics_df], axis=1)

# Save the updated DataFrame back to final_mini.csv
final_mini_df.to_csv(final_mini_file, index=False)

# Print the updated DataFrame
print("\nUpdated final_mini.csv with 2024 Metrics:")
print(final_mini_df[['ticker', '2024_returns', '2024_sharpe_ratio']])
```

```python
import requests
import pandas as pd

# Replace the "demo" API key below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=CACX.LON&outputsize=full&entitlement=delayed&apikey=1JYCUCSEKO7IYQKU'

r = requests.get(url)
data = r.json()
df = pd.DataFrame(data['Time Series (Daily)']).T

# Convert index to datetime
df.index = pd.to_datetime(df.index)

# Convert relevant columns to numeric
df['4. close'] = pd.to_numeric(df['4. close'], errors='coerce')
df['7. dividend amount'] = pd.to_numeric(df['7. dividend amount'], errors='coerce')

# Filter the data for the year 2021
df_2021 = df[df.index.year == 2021]

# Add a 'month' column to group by month
df_2021['month'] = df_2021.index.to_period('M')

# Calculate monthly returns (using only the first and last day of the month)
def calculate_monthly_return(group):
    start_price = group['4. close'].iloc[-1]  # Last close price of the month
    end_price = group['4. close'].iloc[0]    # First close price of the month
    total_dividends = group['7. dividend amount'].sum()  # Total dividends for the month
    monthly_return = (end_price / start_price) - 1
    
    # Debug: Print start price, end price, and total dividends for each month
    print(f"Month: {group['month'].iloc[0]}, Start Price: {start_price:.2f}, End Price: {end_price:.2f}, Total Dividends: {total_dividends:.2f}")
    
    return monthly_return

monthly_returns = df_2021.groupby('month').apply(calculate_monthly_return)

# Convert to percentage and round to 2 decimal places
monthly_returns = (monthly_returns * 100).round(2)

# Print the calculated monthly returns
print("\nMonthly Returns for 2024 (excluding dividends):")
print(monthly_returns)
```

```python
import requests
import pandas as pd

# Replace the "demo" API key below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=CACX.LON&outputsize=full&entitlement=delayed&apikey=1JYCUCSEKO7IYQKU'

r = requests.get(url)
data = r.json()
df = pd.DataFrame(data['Time Series (Daily)']).T

# Convert index to datetime
df.index = pd.to_datetime(df.index)

# Convert relevant columns to numeric
df['4. close'] = pd.to_numeric(df['4. close'], errors='coerce')

# Calculate the absolute and percent change
df['absolute_change'] = df['4. close'].diff()  # Difference between current and previous close
df['percent_change'] = df['4. close'].pct_change() * 100  # Percent change between current and previous close

# Get the most recent price change
most_recent_row = df.iloc[0]  # Most recent row (latest date)
previous_row = df.iloc[1]  # Previous session's row

most_recent_price = most_recent_row['4. close']
previous_price = previous_row['4. close']
absolute_change = most_recent_price - previous_price
percent_change = (absolute_change / previous_price) * 100

# Print the results
print(f"Most Recent Price: {most_recent_price:.2f}")
print(f"Previous Price: {previous_price:.2f}")
print(f"Absolute Change: {absolute_change:.2f}")
print(f"Percent Change: {percent_change:.2f}%")
```
